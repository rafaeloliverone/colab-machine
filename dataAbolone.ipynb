```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import math
import numpy
from collections import Counter

# ========== ABALONE ===========
url_abalone = "https://raw.githubusercontent.com/rafaeloliverone/colab-machine/main/database/abalone.data"


dataset_abalone = pd.read_csv(url_abalone, header=None)
columns_abalone = len(dataset_abalone.columns)

# X and Y 

y = dataset_abalone[0]
X = dataset_abalone.loc[:,1:columns_abalone-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) # 80% treino e 20% teste

print(y_test)
```

Relatório de desempenho dos algoritimos Abalone Resultados

```python
#Árvore 1 entropy
model = tree.DecisionTreeClassifier(criterion="entropy")
model = model.fit(X_train, y_train)
```

```python
# Predição e Resultados

result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```

```python
#Árvore 2 Gini

model = tree.DecisionTreeClassifier(criterion="gini")
model = model.fit(X_train, y_train)
print(y_test)
```

```python
result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```

Resultado kNN

```python
#kNN. euclidean n10
n = 10

model = KNeighborsClassifier(n_neighbors=n, metric='euclidean', algorithm='brute')
model = model.fit(X_train, y_train)

result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```
